```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
df = pd.read_csv("csv_table.csv")
cont = 0
for i in df.time:
  if df.time[cont][-2]=='m':
    df.time[cont]=df.time[cont][:-2]
    df.time[cont]=float(df.time[cont])
  elif df.time[cont][-2]=='µ':
    df.time[cont]=df.time[cont][:-2]
    df.time[cont]=float(df.time[cont])/1000
  else:
    df.time[cont]=df.time[cont][:-1]
    df.time[cont]=float(df.time[cont])*1000
  cont+=1
```

```python
def plotar_grafico(n_threads, n_executions):
  newdf = df[(df.num_threads == n_threads) & (df.executions == n_executions) & (df.lock_type == 'TAS')]
  newdf1 = df[(df.num_threads == n_threads) & (df.executions == n_executions) & (df.lock_type == 'TTAS')]
  plt.title('Desempenho para ' + str(n_threads) + ' threads e ' + str(n_executions) + ' execuções')
  plt.plot(newdf.critical_load, newdf.time, label = 'TAS')
  plt.plot(newdf1.critical_load, newdf1.time, label = 'TTAS')
  plt.legend(loc = 'lower right')

  plt.show()



```

```python
for i in [10,100,1000]:
  for j in [10,100,1000]:
    plotar_grafico(i,j)
```
